```python
# loading librairies
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# loading the IPL dataset
ipl = pd.read_csv('matches.csv')
```

# Exploratory analysis

```python
ipl.head()
```

```python
ipl.shape
```

```python
# most man of the matches
ipl['player_of_match'].value_counts()
```

```python
# top 10  man of matches
ipl['player_of_match'].value_counts()[0:10]
```

```python
# top 5  man of matches
ipl['player_of_match'].value_counts()[0:5]
```

```python
plt.figure(figsize=(8,6))
plt.bar(ipl['player_of_match'].value_counts()[0:5].keys(),ipl['player_of_match'].value_counts()[0:5], color = 'g');
```

```python
# Getting frequency of result column
ipl['result'].value_counts()
```

```python
# Toss wins wrt to each team
ipl['toss_winner'].value_counts()
```

```python
# Where a team won batting first
batting_first = ipl[ipl['win_by_runs']!=0]
batting_first
```

```python
# making histogram
plt.figure(figsize=(5,5))
plt.hist(batting_first['win_by_runs']);
plt.title('Distribution of runs');
plt.xlabel('Runs');
plt.ylabel('Matches');
```

```python
# wins wrt to batting first
batting_first['winner'].value_counts()
```

```python
# making bar plot for top 3 winning teams after batting first
plt.figure(figsize=(7,7))
plt.bar(batting_first['winner'].value_counts()[0:3].keys(), batting_first['winner'].value_counts()[0:3], color =['blue','yellow','orange']);
```
